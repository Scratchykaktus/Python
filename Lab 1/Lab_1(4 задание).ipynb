```python
#Ряд от k = 1 до f < 1e-10, ((-1)^k(x-1)^k)/k^2, 0<=x<=2
#ak = (((-1)^k(x-1)^k)/k^2), ak+1 = (-(-1)^k(x-1)(x-1)^k2)/k(k+1)^2
#ak+1/ak = (-(-1)^k(x-1)(x-1)^k2)/k(k+1)^2 * (k^2/((-1)^k*(x-1)^k)
#ak+1/ak = -(x-1) * (k/k+1)^2

#!/usr/bin/env python3
# -*- coding: utf-8 -*-

import math

if __name__ == "__main__":
    EPS = 1e-10
    x = float(input("Введите значение x, где 0 <= x <= 2"))
    fun = 1
    k = 1
    sum = 0
    if 0 <= x <= 2:
        while fun > EPS:
            fun *= (-(x - 1)) * pow((k/(k+1)), 2)
            sum += fun        
            print("При k = (",k,"), функция f равна = %.10f" % fun)
            k += 1
        print("Sum = %.10f" % sum)
    else:
        print("Ошибка, х должна быть в пределах 0 <= x <= 2")
```
