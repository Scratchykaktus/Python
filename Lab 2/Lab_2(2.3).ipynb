```python
#Дано предложение. Определить, есть ли в нем буквосочетания чу или щу. 
#В случае положительного ответа найти также порядковый номер первой буквы 
#первого из них.

#!/usr/bin/env python3
# -*- coding: utf-8 -*-

if __name__ == '__main__':

    text = 'Я помню чудное мгновенье щу'

    if text.find('чу') != -1:
        print(f"Порядковый номер чу:{text.find('чу')}")
    else:
        print("Буквосочетание 'чу' в тексте нет!")
    if text.find('щу') != -1:
        print(f"Порядковый номер щу:{text.find('щу')}")
    else:
        print("Буквосочетание 'щу' в тексте нет!")

            
```

```python
#Дано слово, оканчивающее символом «.». 
#Составить программу, которая вставляет некоторую заданную букву 
#после буквы с заданным номером.

#!/usr/bin/env python3
# -*- coding: utf-8 -*-

if __name__ == '__main__':

    text = 'Привет, мир!'
    print(f"Исходный текст: {text}")
    s = input("Введите букву, которую хотите вставить: ")
    v = int(input("Введите номер места, после которого хотите вставить букву: "))
    text = text[:v] + s + text[v:]
    print(text)
```
