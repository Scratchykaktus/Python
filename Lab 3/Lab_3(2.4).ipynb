```python
#Список целых чисел, найти минимальный по модулю элемент, сумма модулей, после
#первого нуля, преобразовать список, что первая половина были четные позиции, вторая не четные

#!/usr/bin/env python3
# -*- coding: utf-8 -*-

import math

if __name__ == '__main__':
    list_int_number = [4, 3, 7, 0, -4, -1, 1, 14, -9]
    min = abs(list_int_number[0])
    for i in range(len(list_int_number)-1):
        if min > abs(list_int_number[i+1]):
            min = abs(list_int_number[i+1])
            
    print(f"Минимальный по модулю элемент равен: {min}")

    null_index = list_int_number.index(0)
    list_after_null = list_int_number[null_index:]
    sum = 0
    for i in range(len(list_after_null)):
        sum += abs(list_after_null[i])

    print(f"Сумма модулей элементов после первого нуля равна: {sum}")

    print(f"Исходный список: {list_int_number}")
    list_chet = list_int_number[1::2]
    list_nechet = list_int_number[::2]
    new_list = list_chet + list_nechet
    print(f"Преобразованный список:",new_list)


```
